```python
def stat_parsing(problem, num_trial=1000, parser=parse):
    def _stat(x):
        return np.mean(x), (np.std(x) * 100 // 1) / 100, np.min(x), np.max(x)
    
    n_numbers, n_variables, n_formulas, n_equations, n_lists = [], [], [], [], []
    for i in range(num_trial):
        values, question = problem.random_question()
        objects, numbers, variables, formulas, equations, lists = parser(question)
        n_numbers.append(len(numbers))
        n_variables.append(len(variables))
        n_formulas.append(len(formulas))
        n_equations.append(len(equations))
        n_lists.append(len(lists))
    return _stat(n_numbers), _stat(n_variables), _stat(n_formulas), _stat(n_equations), _stat(n_lists)

        
```

```python
_names = ['n', 'v', 'f', 'e', 'l']

for prob in Problems:
    print('****************', prob.__name__, '******************' )
    stats = stat_parsing(prob, num_trial=1000)
    for i, n in enumerate(_names):
        print(n, stats[i])
```
